<h1>Oblasti platnosti proměnných</h1>

- Typ oblasti platnosti proměnných udává části programu, ve kterých je možné proměnnou získat.
- Rozlišujeme tyto oblasti:
    - globální: Proměnná je dostupná v celém kódu
    - lokální: Proměnná je dostupná pouze v rámci dané části skriptu
- Pořadí, ve kterém Python načítá proměnné, je následující:
    - lokální proměnné
    - přiložené proměnné
    - globální proměnné
    - vestavěné (systémové) proměnné
    
   

```python
mesto = 'Praha'

def vypis_mesto():
    print(mesto)
    
vypis_mesto()



def vypis_mesto2():
    mesto = 'Brno'
    print(mesto)

vypis_mesto2()
print(f'Hodnota promenne mesto je: {mesto}')



def vypis_mesto3():
    #dochazi ke zmene globalni promenne
    global mesto
    mesto = 'Brno'
    print(mesto)

    
vypis_mesto3()
print(f'Hodnota promenne mesto je: {mesto}')
     
```

```python
mesto = 'Praha'
def vypis_mesto4():
    mesto = 'Brno'
    def vypis():
        # Nastavujeme upravu pro promennou v prilozene oblasti
        nonlocal mesto
        mesto = 'Ostrava'
        print(f'1. {mesto}')
    vypis()
    print(f'2. {mesto}')

    
vypis_mesto4()
print(f'3. {mesto}')
```
